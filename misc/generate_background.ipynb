# コースに対応する背景画像の生成

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# GB画面サイズ
w = 160
h = 144
```

```python
# 本来のコース
pos = np.linspace(-1.2,0.6)
plt.plot(pos,np.sin(3*pos))
```

# この画像をGBの背景用に作成

横幅160px -> -1.2~0.6

高さ方向について
- 頂点に旗の絵を足したいので、y座標1.00より上に16pxあけたい
- 下に2行分(16px)文字表示領域を残したい
- 144 - 16 - 16 = 108pxの領域にコースを表示する
- 108px -> -1.0 ~ 1.0

```python
px_x = np.arange(w)
x = (px_x / w) * (0.6 - -1.2) + (-1.2)
```

```python
x
```

```python
y = np.sin(x * 3)
```

```python
px_y = 108 + 16 -  (y + 1) / 2 * 108
```

```python
px_y
```

```python
img = np.zeros((h, w), dtype=np.uint8)
```

```python
img[:] = 255
for i in range(w):
    img[int(px_y[i])-1:int(px_y[i])+2,i]=0
```

```python
plt.imshow(img)
```

```python
cv2.imwrite("../gb/bg.png", img)
# このあと手動で編集する
```
